```python
import torch
from transformers import AutoModel, AutoTokenizer
from google.colab import files

model = AutoModel.from_pretrained("ibm/MoLFormer-XL-both-10pct", deterministic_eval=True, trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("ibm/MoLFormer-XL-both-10pct", trust_remote_code=True)
```

```python
import pandas as pd
import numpy

valid2_df = "valie2"
test_df = "test"

valid2_smiles = pd.read_csv(valid2_df + ".csv")['SMILES'].tolist()
test_smiles = pd.read_csv(test_df + ".csv")['SMILES'].tolist()
```

```python
inputs = tokenizer(valid2_smiles, padding='max_length', truncation=True, return_tensors="pt", max_length = 512)
with torch.no_grad():
    outputs = model(**inputs)
outputs.pooler_output
```

```python
# check the dimension of outputs.pooler_output
outputs.pooler_output.size() # 356 * 768
len(valid2_smiles)
```

```python
import pandas as pd

# Step 2: Convert the tensor to a pandas DataFrame
df = pd.DataFrame(outputs.pooler_output)

df.insert(0, 'smiles', valid2_smiles)

# Step 3: Rename the columns (adjust this step accordingly)
column_names = ['SMILES'] + [f'molformer_feature_{i+1}' for i in range(outputs.pooler_output.shape[1])]
df.columns = column_names

# Step 4: Save the DataFrame to a CSV file
df.to_csv(f'molformer_{valid2_df}_features.csv', index=False)
files.download(f'molformer_{valid2_df}_features.csv')
```

```python
inputs = tokenizer(test_smiles, padding='max_length', truncation=True, return_tensors="pt", max_length = 512)
with torch.no_grad():
    outputs = model(**inputs)
outputs.pooler_output

df = pd.DataFrame(outputs.pooler_output)
df.insert(0, 'smiles', test_smiles)

# Step 3: Rename the columns
column_names = ['SMILES'] + [f'molformer_feature_{i+1}' for i in range(outputs.pooler_output.shape[1])]
df.columns = column_names

# Step 4: Save the DataFrame to a CSV file
df.to_csv(f'molformer_{test_df}_features.csv', index=False)
files.download(f'molformer_{test_df}_features.csv')
```
